```python
data = pd.read_csv('https://raw.githubusercontent.com/Defcon27/Data-Analysis-of-Indian-Automobile-dataset-using-Machine-Learning-in-R/master/indian-auto-mpg.csv', index_col=0)
data.head()
```

```python
import matplotlib.pyplot as plt
s = data.groupby('Fuel_Type')['Price'].agg("mean")
y = list(s)
mylabels = s.index
plt.pie(y, labels = mylabels)
plt.title('fuel type prices')
plt.show() 
```

```python
import seaborn as sns
corr = data.corr()
fig, ax = plt.subplots(figsize=(10,10))                                 # plotting co-relation matrix
sns.heatmap(corr,annot=True, cmap = "Blues", linewidth = 0.30)
plt.title("Correlation matrix of features")
plt.show()
```
